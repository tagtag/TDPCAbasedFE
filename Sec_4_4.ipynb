Table 4.6の結果を出します。

```r
pca0 <- NULL
reg0<-NULL
for (i in c(1:100))
{
cat(i," ")
x <- rnorm(10)
x <- rep(x,5)
x<- matrix(x[apply(matrix(1:10,ncol=1),1,function(i){rep(1:50,2)[i:(i+49)]})],nrow=50)
x0 <- matrix(rnorm(50*100),nrow=50)*0.1
x0[,1:10] <- x+x0[,1:10]
x<- x0
pca <- prcomp(t(x))
pca0 <- c(pca0,sum(order(-rowSums(pca$x[,1:2]^2))[1:10] %in% 1:10))
a <- sin(c(1:50)/10*2*pi)
b <- cos(c(1:50)/10*2*pi)
LM <- lm(x~a+b)
SLM <- summary(LM)
reg0 <-c(reg0,sum(order(-abs(unlist(lapply(SLM,"[",8))))[1:10] %in% 1:10))
}
mean(pca0)
mean(reg0)
```
